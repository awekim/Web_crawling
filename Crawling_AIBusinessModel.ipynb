```python
import selenium
from selenium import webdriver 
from selenium.webdriver.chrome.service import Service #chromedriver 자동설치
from webdriver_manager.chrome import ChromeDriverManager #chromedriver 자동설치
from selenium.webdriver.common.by import By #요소 찾을 때 사용

import pandas as pd
```

```python
service = Service(executable_path=ChromeDriverManager().install()) #크롬드라이버 설치
driver = webdriver.Chrome(service=service)
```

```python
driver.get('https://www.innoforest.co.kr/dataroom?id=CP00003304&page=1&tagArray=TG00070016&tagText=%EC%83%9D%EC%84%B1%ED%98%95AI')
```

### Page 1

```python
# 다음으로 누르기
driver.find_element(By.CSS_SELECTOR, '#__next > div.css-hmxjkf > div > div > button').click()
driver.find_element(By.CSS_SELECTOR, '#__next > div.css-hmxjkf > div > div > button').click()
```

```python
driver.find_element(By.CSS_SELECTOR, '#__next > main > div > div > div.common-modal-header.css-27ru1e > div > span > img').click()
#driver.find_element(By.CSS_SELECTOR, '#__next > main > div > div > div.common-modal-footer.css-1y4cmsu > button.css-1g6cx0i').click()
```

```python
# css-8sdmb class 찾기
table = driver.find_element(By.CSS_SELECTOR, '#__next > main > div.css-c0cixj > div:nth-child(2) > div > article > table')                
```

```python
data = []

for tr in table.find_elements(By.TAG_NAME, 'tr'):
    row = []
    for td in tr.find_elements(By.TAG_NAME, 'td'):
        row.append(td.text)
    data.append(row)
    
# data를 data.frame으로 만들기
df = pd.DataFrame(data)
df
```

```python
df.to_csv('AI_BM_01.csv', encoding='utf-8-sig')
```
